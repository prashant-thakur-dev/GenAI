### structure outopt

```python
from pydantic import BaseModel, Field
from typing import List

class ResponseStructure(BaseModel):
    name: str = Field(description="Complect Name")
    email: str = Field(description="Email Address")
    age: int = Field(description="Age")

structured_llm = gemini_llm.with_structured_output(ResponseStructure)
```

```python
res = structured_llm.invoke(f"please give me only important content: {text}")
res
```

```python
res.model_dump()
```

```python
class Movies(BaseModel):
    title: str = Field(description="Movie Title")
    year: int = Field(description="Movie Release Year")

class AllMovies(BaseModel):
    movies: List[Movies]

movie_llm = gemini_llm.with_structured_output(AllMovies)
```

```python
res = movie_llm.invoke("give me 10 recient movie")
res
```
