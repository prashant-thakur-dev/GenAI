```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
prompts = [
    {"role":"user", "content": "my hame is raj"},
    {"role":"ai", "content": "Hi Raj, it's nice to meet you! How can I help you today?"},
    {"role":"user", "content": "what is my name"},
    
]

res = llm.invoke(prompts)
```

```python
res.content
```

### manage history


```python
history = []
while True:
    query = input("User: ")
    history.append({"role":"user", "content": query})
    print("User: ", query)

    if query in ["exit", "quit", "q"]:
        print("bye")
        break

    res = llm.invoke(history)
    history.append({"role":"ai", "content": res.content})
    print(res.content)
```
